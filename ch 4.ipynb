## PYTHON BASED METHOD


```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

#### second Step
- import dataset

```python
x=np.array([60,62,67,70,71,72,75,78]).reshape(-1,1)
```

```python
print(x)
```

```python
y=np.array([140,155,159,179,192,200,212,214])
print(y)
```

## third step
- create model

```python
model=LinearRegression()
```

## fourth step

```python
model.fit(x,y)
```

```python
#fifth step 
# print the slope and intercept
# slope
print(model.coef_[0])
print(model.intercept_)
```

```python
slope=model.coef_[0]
intercept=model.intercept_
print(f"the linear regression equation is y={intercept:.2f}+{slope:.2f}*x")
```
