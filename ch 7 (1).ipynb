## ACCESSING NASA API

```python
import requests
import os

API_KEY=" "
APOD_URL='https://api.nasa.gov/plantary/apod'
# create directory
os.makedirs("Nasa_images",exist_ok=True)
params ={
    'api_key' : API_KEY,
    'count' :3
}
response = requests.get(APOD_URL,params=params)
apodimages = response.json()
for i,apod in enumerate(apodimages):
    imageurl =apod.get('hdurl',apod.get('url'))
    if imageurl and imageurl.endswith(('jpg','png')):
        imagepath=os.path.join("Nasa images",f'apod_{i+1}.jpg')
        print(f"downloading image {i+1}")
        imageresponse = requests.get(imageurl)
        with open(imagepath,'wb') as file:
            file.write(imageresponse.content)
        print(f"saved to the {imagepath}")
    else:
        print(f"skipping non image")
        
```

## API METHODS

### `GET METHOD`
- This method is used to fetch the api resource content

```python
import requests
import json
import pandas as pd
url='http://jsonplaceholder.typicode.com/posts'
response=requests.get(url)
print(response)
print(response.status_code)
print(response.text)
print(type(response.text))
print(response.json())
print(type(response.json()))
```

## `GET METHOD` parameters
- parameters
- headers

```python
params={
    'userID': 1
}
url='http://jsonplaceholder.typicode.com/posts'
response=requests.get(url,params=params)
pprint.pprint(response.json())
```

## headers

```python
headers={
    "User-Agent":"Mozilla/5.0",
    "Accept":"application/json"
}
response =requests.get(url,headers=headers)
print(response.status_code)
```

## `POST METHOD`
- This method is use to create a new resource

```python
new_post={
    "tittle":"Learning API",
    "body": "This is a test content",
    "userId" :1
}
response = requests.post("https://jsonplaceholder.typicode.com/posts/1",
                         json = new_post)
print(response)
print(response.json())
```

## `PUT METHOD`
- This methods is used to update the whole resource

```python
updated_post ={
    "id" :1,
    "title":"updated title",
    "body" :"updated body",
    "userId" : 1
}
response = requests.put("https://jsonplaceholder.typicode.com/posts/1",
                         json = updated_post)
print(response)
print(response.json())
```

## `PATCH METHOD`
- this method is used to partially update the resorce

```python
patched_data={
    'title' :"patched_title"
}
response= requests.patch("https://jsonplaceholder.typicode.com/posts/1",
                        json = patched_data)
print(response)
print(response.json())
```

## `DELETES METHOD`
- THIS METHOD DELETES THE RESOURSE

```python
response= requests.patch("https://jsonplaceholder.typicode.com/posts/1")
print(response)
print(response.json())
```

## `PAGINATION`
- this method is feature allows paginated API access to the user

```python
import requests
import json
import pandas as pd
url='http://jsonplaceholder.typicode.com/posts'
page = 1
while page <=3:
    params ={
        "_page":1,
        "_limit":5
    }
    response=requests.get(url,params=params)
    posts=response.json()
    print(f"\nPage{page}")
    for post in posts:
        print(post['id'],post['title'])
    page+=1
```
