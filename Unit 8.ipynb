#### 2. create app
```python
python manage.py startapp findmusic
- register app in 'musicplayer/settings.py'
```

#### 3. models.py
```python
from django.db import models
class Music(models.Model):
   song=models.charfiled(max_length=100)
   artist=models.charfiled(max_length=100)
   album=models.charfiled(max_length=100)
   year=models.IntegerField()
   def __str(self):
      return self.song
```
#### optional (register model in admin) 
```python
from .models import Music
admin.site.register(Music)
```
#### 4. Run migration and create superuser
```python
python manage.py makemigrations
python manage.py migrate
python manage.py cretessuperuser
```
#### 5. view.py
```python
from django.shortcut import render
from .models import Music

def findmusic(request):
    musiclist =Music.objects.all()
    context ={
        'musiclist':musiclist
    }
    return render(request,'home.html',context)
```
