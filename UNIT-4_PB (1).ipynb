### 215

For x = np.array([5, 15, 25, 35, 45, 55]) and y = np.array([5, 20, 14, 32, 22, 38]), apply simple linear regression using scikit learn library and calculate calculate R squared, coeficient and intercept. Predict the y values for x = np.arange(5). (Don't split data for training/testing)

---

```python
#import the libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
# define features and target
x = np.array([5, 15, 25, 35, 45, 55]).reshape(-1,1)
y = np.array([5, 20, 14, 32, 22, 38])
print(x.shape)
print(y.shape)
# create model
model=LinearRegression()
# fit the model
model.fit(x,y)
# print r2,score slope and intercept 
print(model.score(x,y))
slope=model.coef_[0]
intercept=model.intercept_
print(f"Regression Line equation is y={slope:.2f}*x+{intercept:.2f}")
#predict for np.arange(5)
model.predict(np.arange(5).reshape(-1, 1))
```

### 217

Given a real estate price size year dataset, implement multiple linear regression using scikitlearn library. Using the model, make a prediction about an apartment price with size 750 sq.ft. for 2009.Also Calculate R squared, coeficient and intercept. (Don't split data for training/testing)

---

```python
# multiple linear resgesssion 
import pandas as pd
data=pd.DataFrame({
    'size':[2104,1600,2400,1416,3000],
    'year':[2003,1977,2001,1995,2000],
    'price':[399900,329900,369000,232000,539900]
})
x=data[["size","year"]]
y=data["price"]
print(x.shape)
print(y.shape)
model=LinearRegression()
model.fit(x,y)
# print r2,score slope and intercept b
print(model.score(x,y))
print("the slope is ",model.coef_[0])
print("the intercept is",model.intercept_)
newdata=pd.DataFrame({
    'size':[750],
    'year':[2009]
})
prediction=model.predict(newdata)
print("predication for 750 sqFt house with 2009 year is",prediction[0])
```
